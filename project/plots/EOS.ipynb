```python
dirr = '/home/hugens/shared/uni/modsim/modsim/project/backup/EOS/'
os.chdir(dirr)
betaPv0s = [];
avg_rhos = [];
for file in os.listdir(os.getcwd()):
    f = open(file)
    line = f.readline()
    betaPv0 = int(line.split("\n")[0])
    betaPv0s.append(betaPv0)
    steps, rhos = np.loadtxt(file, skiprows=1).transpose()
    avg_rhos.append(np.mean(rhos))
    plt.plot(steps, rhos, label="b = " + str(betaPv0))
    plt.xlabel('steps')
    plt.ylabel('rho')
#     plt.show()
    plt.legend()
plt.show()
plt.plot(betaPv0s, avg_rhos)
```
