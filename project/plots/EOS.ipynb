```python
dirr = '/home/hugens/shared/uni/modsim/modsim/project/backup/EOS/'
os.chdir(dirr)
betaPv0s = [];
avg_rhos = [];
avg_Ss   = [];
for file in os.listdir(os.getcwd()):
    f = open(file)
    line = f.readline()
    betaPv0 = int(line.split("\n")[0])
    betaPv0s.append(betaPv0)
    steps, rhos, Ss = np.loadtxt(file, skiprows=1).transpose()
    avg_rhos.append(np.mean(rhos))
    avg_Ss.append(np.mean(Ss))
    plt.plot(steps, rhos, label="b = " + str(betaPv0))
    plt.xlabel('steps')
    plt.ylabel('rho')
#     plt.show()
    plt.legend()
plt.show()
plt.plot(avg_rhos, betaPv0s, '.')
plt.show()
plt.plot(avg_rhos, avg_Ss, '.')
```
