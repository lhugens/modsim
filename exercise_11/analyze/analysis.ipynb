```python
import numpy as np
import matplotlib.pyplot as plt
import os
from time import sleep
```

```python
fig = plt.figure(figsize=(10,5))
os.chdir(os.getcwd()+"/../hist")
dic = {'000':'b', '001':'r', '002':'k'}
mgs = ['000', '001', '002']

for mg in mgs:
    avg = np.zeros(3)
    max_zs = np.zeros(1)
    for file in os.listdir(os.getcwd()):
        if ('mg' + mg) in file:
            mg_i = file[7:7+3]
            try:
                zs, values = np.loadtxt(file).transpose()
                if len(zs) > len(max_zs):
                    max_zs = np.copy(zs)
                if len(avg) > len(values):
                    values = np.hstack((values, np.zeros(len(avg)-len(values))))
                elif len(avg) < len(values):
                    avg = np.hstack((avg, np.zeros(len(values)-len(avg))))
                avg = avg + values
            except:
                break
    try:
        plt.plot(max_zs, avg/np.sum(avg), dic[mg]+'.')
        plt.plot(max_zs, avg/np.sum(avg), dic[mg]+'-', linewidth=0.5)
    except:
        continue
plt.ylabel(r'$\rho(z)$')
plt.xlabel(r'$z$')
```
